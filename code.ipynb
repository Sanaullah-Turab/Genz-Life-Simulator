```python
# Function that matches user input to appropriate responses and detects trigger keywords for special features
def GetResponse(user_input):
    user_input = user_input.lower().strip()
    
    if user_input in responses:
        return responses[user_input]
    elif "play" in user_input or "game" in user_input:
        return "game_trigger"
    elif "cafe" in user_input or "drink" in user_input or "coffee" in user_input:
        return "cafe_trigger"
    elif "music" in user_input or "song" in user_input:
        return "music_trigger"
    else:
        return "Hmm... idk what to say ğŸ¤”"
```

```python
# Function that displays current player statistics including energy, points, and mood
def ShowStatus():
    print("\n" + "â”€"*40)
    print("ğŸ“Š YOUR CURRENT STATS")
    print("â”€"*40)
    print(f"âš¡ Energy Level: {player_stats['energy']}/100 (we're vibing!)")
    print(f"ğŸ† Points Earned: {player_stats['points']} (you're slaying!)")
    print(f"ğŸ˜Š Current Mood: {player_stats['mood']} (the energy is real!)")
    print("â”€"*40 + "\n")
```

```python
# Function that updates player statistics based on different activities (study, eat, play, sleep)
def UpdateStats(activity):
    if activity == "study":
        player_stats["energy"] -= 20
        player_stats["points"] += 10
        player_stats["mood"] = "ğŸ˜¤"
        print("\nğŸ“š Time to hit the books bestie! Brain gains incoming!")
        time.sleep(0.8)
        print("Energy: -20 | Points: +10 | Mood: Productive grind mode ğŸ˜¤")
    elif activity == "eat":
        player_stats["energy"] += 15
        player_stats["mood"] = "ğŸ˜‹"
        print("\nğŸ• Snack time! Fueling up like the icon you are!")
        time.sleep(0.8)
        print("Energy: +15 | Mood: Food coma incoming ğŸ˜‹")
    elif activity == "play":
        player_stats["energy"] -= 10
        player_stats["points"] += 5
        player_stats["mood"] = "ğŸ˜"
        print("\nğŸ® Gaming mode activated! Solo Leveling vibes fr!")
        time.sleep(0.8)
        print("Energy: -10 | Points: +5 | Mood: Gamer vibes activated ğŸ˜")
    elif activity == "sleep":
        player_stats["energy"] = 100
        player_stats["mood"] = "ğŸ˜´"
        print("\nğŸ˜´ Beauty sleep time! Recharging those main character batteries!")
        time.sleep(1.2)  # Longer pause for sleep effect
        print("Energy: Fully Restored to 100 | Mood: Well rested queen âœ¨ğŸ˜´")
    # Ensure energy stays within bounds (0-100)
    player_stats["energy"] = max(0, min(100, player_stats["energy"]))
```

```python
# Unique feature #1: Song recommendation system that suggests music based on current mood and awards points
def RecommendSong():
    global playlist_score
    current_mood = player_stats["mood"]
    
    print("\nğŸµ MOOD BOOSTER ACTIVATED! ğŸµ")
    print("â”€"*40)
    
    # Get playlist for current mood, with fallback to default
    playlist = mood_playlists.get(current_mood, ["Good Vibes Mix", "Chill Hits", "Mood Mix"])
    song = random.choice(playlist)
    
    if current_mood in mood_playlists:
        print(f"Based on your mood {current_mood}, here's a vibe:")
        print(f"ğŸ¶ Now playing: {song}")
        
        mood_boost = random.randint(5, 15)
        player_stats["points"] += mood_boost
        playlist_score += 10
        
        print(f"\nâœ¨ Mood lifted! +{mood_boost} points")
        print(f"ğŸµ Playlist Score: {playlist_score}")
    else:
        print(f"ğŸ¶ Playing: {song}")
        player_stats["points"] += 5
        playlist_score += 5
    
    print("â”€"*40 + "\n")
```

```python
# Unique feature #2: Virtual cafe with drink menu and random spill chance that affects energy and mood
def MiniCafe():
    # Ensure player has a valid mood
    if not player_stats["mood"]:
        player_stats["mood"] = "ğŸ˜Š"
    
    print("\nâ˜• WELCOME TO BESTIE'S CAFE! â˜•")
    print("â”€"*40)
    print("What would you like to order?")
    print()
    
    for idx, (drink, details) in enumerate(cafe_menu.items(), 1):
        print(f"{idx}. {details['price']} {drink} (Energy: +{details['energy']}, Mood: {details['mood']})")
    
    print("0. Leave Cafe")
    print("â”€"*40)
    
    try:
        choice = int(input("Choose a drink (0-5): "))
        
        if choice == 0:
            print("Come back soon! â˜•")
            return
        elif 1 <= choice <= len(cafe_menu):
            drink_name = list(cafe_menu.keys())[choice - 1]
            drink_details = cafe_menu[drink_name]
            
            spill_chance = random.randint(1, 10)
            
            if spill_chance == 1:
                print(f"\nğŸ˜± OH NO! Your {drink_name} got spilled!")
                print("ğŸ’” -5 energy from the disappointment...")
                player_stats["energy"] -= 5
                player_stats["mood"] = "ğŸ˜¢"
            else:
                print(f"\nâ˜• You sipped a {drink_name}!")
                player_stats["energy"] += drink_details["energy"]
                player_stats["mood"] = drink_details["mood"]
                print(f"âœ¨ +{drink_details['energy']} energy, mood: {drink_details['mood']}")
            
            # Ensure energy stays within bounds
            player_stats["energy"] = max(0, min(100, player_stats["energy"]))
        else:
            print("Invalid choice! Try again bestie ğŸ¤·")
    except:
        print("Oops! Invalid input bestie ğŸ˜…")
```

```python
# Main game loop function that displays activity menu and handles user choices for life simulation
def PlayGame():
    global player_stats
    
    while True:
        ShowStatus()
        print("\nğŸ® ACTIVITY MENU ğŸ®")
        print("â”€"*40)
        print("1. ğŸ“š Study (Energy -15, Knowledge +10)")
        print("2. ğŸ• Eat (Energy +20, Mood boost)")
        print("3. ğŸ‰ Play (Energy -10, Mood boost)")
        print("4. ğŸ˜´ Sleep (Energy +40, Rest up)")
        print("5. â˜• Visit Cafe (Random vibes!)")
        print("6. ğŸµ Get Song Recommendation (Mood boost)")
        print("0. âŒ Quit Game")
        print("â”€"*40)
        
        try:
            choice_input = input("\nWhat's the vibe? ").strip()
            choice = int(choice_input)
            
            if choice == 0:
                EndGame()
                break
            elif choice == 1:
                print("\nğŸ“š You're studying hard! Knowledge gaining... ğŸ§ ")
                UpdateStats("study")
            elif choice == 2:
                print("\nğŸ• Nom nom nom! Energy restored! ğŸ”")
                UpdateStats("eat")
            elif choice == 3:
                print("\nğŸ‰ Let's goooo! Party time! ğŸŠ")
                UpdateStats("play")
            elif choice == 4:
                print("\nğŸ˜´ Sweet dreams bestie... Zzz... ğŸ’¤")
                UpdateStats("sleep")
            elif choice == 5:
                MiniCafe()
            elif choice == 6:
                RecommendSong()
            else:
                print("\nâŒ Invalid choice! Try again bestie ğŸ¤¦")
        except ValueError:
            print("\nâš ï¸ Oops! Please enter a number (0-6)! Try again ğŸ˜…")
        except Exception as e:
            print(f"\nâš ï¸ Something went wrong: {e}. Try again ğŸ˜…")
```

```python
# Main conversation loop that handles chat interaction and triggers special features based on keywords
def ChatLoop():
    global player_stats
    
    print("\nğŸ’¬ Let's chat! (Type 'bye' to end chat)\n")
    
    while True:
        try:
            user_input = input(f"{user['name']}: ").strip().lower()
            
            if not user_input:
                continue
        except KeyError:
            print("âš ï¸ User profile not found! Please restart.")
            return
        
        response = GetResponse(user_input)
        time.sleep(0.8)  # Simulate typing delay
        print(f"Whisper: {response}")
        
        game_trigger = ["game", "play", "activities", "do something", "bored"]
        cafe_trigger = ["cafe", "coffee", "drink", "thirsty", "cafe time"]
        music_trigger = ["song", "music", "playlist", "recommend", "listen"]
        
        if any(word in user_input for word in game_trigger):
            time.sleep(0.5)
            print("\nğŸ® Ready to start the life simulator? Let's go!")
            time.sleep(1)
            PlayGame()
            break
        elif any(word in user_input for word in cafe_trigger):
            MiniCafe()
        elif any(word in user_input for word in music_trigger):
            RecommendSong()
        elif user_input in ["bye", "goodbye", "exit", "quit", "stop"]:
            EndChat()
            break
```

```python
# Main function that orchestrates the entire program by calling GreetUser and ChatLoop
def main():
    GreetUser()
    ChatLoop()

if __name__ == "__main__":
    main()
```
